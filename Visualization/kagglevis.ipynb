<B>Specific Case Example
For 'Avila Adobe'</b></br>Checking if visitor rate seasonally and when the museum should hire additional season staff

```python
plt.figure(figsize=(16,4))

plt.title("Avila Adobe Seasonal Spike")

sns.lineplot(data = museum_data['Avila Adobe'])

plt.xlabel = "Date"
```

<H4>The seasons are:</H4><br>
September-February (in LA, the fall and winter months) <br>
in March-August (in LA, the spring and summer) <br>

The chart generally dips to the mimimum during beginning and ends of a year (in December and January) while it peaks during the middle of <br> the year (especially around May and June). Thus, Avila Adobe gets more visitors during March-August Period. <br>
With this in mind, seasonal employees should be hired during the spring and summer seasons by Avila Adobe

<h3> Bargraphs and Heatmaps <h3>

```python
ign = "../Data/ign_scores.csv"
flight = "../Data/flight_delays.csv"

ign_data = pd.read_csv(ign, index_col='Platform')
flight_data = pd.read_csv(flight, index_col='Month')
```

```python
ign_data
```

```python
flight_data
```

```python
#bargraphs

plt.figure(figsize=(25,5))

plt.title("Average Game Score for 'Fighting' genre by Platform")

sns.barplot(x = ign_data.index , y = ign_data.Fighting)
```

```python
plt.figure(figsize=(24,4))
plt.title("Average Delays per month for OO line")
sns.barplot(x = flight_data.index, y = flight_data.OO)
```

```python
sns.barplot(x = flight_data.index, y = flight_data.UA)
```

```python
#Heatmaps

plt.figure(figsize=(16,7))

plt.title("Heatmap Score of all Genres of Videogames by Platform")

sns.heatmap(data = ign_data, annot = True)
```

```python
plt.figure(figsize=(16,5))

plt.title("Flight delays by months for all airlines")

sns.heatmap(data= flight_data, annot = True)
plt.xlabel = "Airlines" 
```

```python
#Seaborn's heatmap generates its own internal axes object which overwrites whatever plt.xlabel() tries to set. To solve this
ax = sns.heatmap(data=flight_data, annot = True)
ax.set_xlabel("Airlines")
ax.set_ylabel("Months")
```

```python
ax = sns.heatmap(data = ign_data, annot=False)
ax.set_xlabel("Genres")
```

<h3>Distributions</h3> <br> <h4>Histograms and Density Plots</h4>

```python
iris_path = "../Data/iris.csv"
iris_data = pd.read_csv(iris_path)
iris_data
```

```python
#Histograms
#see how petal length varies in iris flowers
sns.histplot(data=iris_data['Petal Width (cm)'])
```

```python
#density plots /Kernel density estimate (KDE)
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
#2D KDE plots //
# In the plot below, the color-coding shows us how likely we are to see different 
# combinations of sepal width and petal length, where darker parts of the figure are more likely

sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Length (cm)'], kind='kde')
```

```python
#color coded plots
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species' )
```

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)
```
