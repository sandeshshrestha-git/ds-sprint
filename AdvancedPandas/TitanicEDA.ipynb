```python
#survival rate by sex
Total = train.groupby('Sex')['Survived'].count() #it counts the total no. of both genders bc, both 0 and 1 are not null values
Total
```

```python
survived = train.query('Survived == 1').groupby('Sex')['Survived'].count() #this only counts the values of '1' in 'Survived' thus giving the no. of only survived people in each gender
rate_of_survival = survived / Total
rate_of_survival
```

```python
#survival rate by class
PclassTotal = train.groupby('Pclass')['Survived'].count()
PclassTotal
```

```python
Pclass_survived = train.query('Survived == 1').groupby('Pclass')['Survived'].count()
Pclass_survived
```

```python
rate_by_Pclass = Pclass_survived / PclassTotal
rate_by_Pclass
```

```python
#avg age of survivors vs non survivors
avg_survivors = train.query('Survived == 1').groupby('Survived')['Age'].mean()
avg_survivors
```

```python
avg_died = train.query('Survived == 0').groupby('Survived')['Age'].mean()
avg_died
```

```python
avg_age = train.groupby('Survived').Age.mean()
avg_age
```

```python
#which embarkment point had the highest survival rate
total_embarkment = train.groupby('Embarked')['Survived'].count()
total_embarkment
```

```python
survived_by_embarkment = train.query('Survived == 1').groupby('Embarked')['Survived'].count()
survived_by_embarkment
```

```python
sorted_survival_by_embark = (survived_by_embarkment / total_embarkment).sort_values(ascending = False)
```

```python
sorted_survival_by_embark
```

```python
#survival when travelling alone vs with family
total_family_survived = train.query('Survived == 1').groupby('Survived')['newCol'].count()
total_family_survived
```

```python
solo_survived = train.query('newCol == 1 and Survived == 1').groupby('Survived')['newCol'].count()
solo_survived
```

```python
family_survived = train.query('newCol > 1 and Survived == 1').groupby('Survived')['newCol'].count()
family_survived
```

```python
solo_survival_rate = solo_survived / total_family_survived
solo_survival_rate
```

```python
family_survived_rate = family_survived / total_family_survived  
family_survived_rate
```

```python
family_survived_rate >solo_survival_rate
```
