Creating a Series for ingredients 

Flour     4 cups  <br>
Milk       1 cup<br>
Eggs     2 large<br>
Spam       1 can<br>
Name: Dinner, dtype: object

```python
ingredients = pd.Series(['4 cups','1 cup','2 large','1 can'],
                        index=['Flour','Milk','Eggs','Spam'],
                        name='Dinner', dtype='object')
ingredients
```
